```python
#Importando Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/corrida.png')
```

```python
#Convers√£o de BGR to RGB
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Convers√£o das Imagens para escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Carrega-se o Descritor de HOG
hog = cv2.HOGDescriptor()
```

```python
#Para a Detec√ß√£o de Pessoas com HOG, ser√° utilizado SVM
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
#Criando as caixas de Detec√ß√£o para as Imagens
boxes, weights = hog.detectMultiScale(img_gray)
```

```python
#Inser√ß√£o das Caixas na Imagem Colorida, ap√≥s a detec√ß√£o do Descritor
for (x, y, w, h) in boxes:

    #Defini√ß√£o da Esquerda Superior, Direita Inferior, Cor e Largura da Caixa
    pontoEsqSuperior = (x, y)
    pontoDirInferior = (x + w, y + h)
    cor = (0, 255, 0)
    espessura = 2

    #Inser√ß√£o das Caixas
    cv2.rectangle(img, pontoEsqSuperior, pontoDirInferior, cor, espessura)

plot.imshow(img,cmap='gray')
plot.show()
```
