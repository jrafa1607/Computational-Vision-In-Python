```python
#Import de Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/objetos.bmp')
```

```python
#Visualização da Imagem Importada
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Inicialmente, converte-se a imagem colorida para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Em seguida, aplica-se a Limiarização
T = 50
T ,img_bin = cv2.threshold(img_gray, T, 255, cv2.THRESH_BINARY)
```

```python
#Visualização da Imagem Binarizada
plot.imshow(img_bin,cmap='gray')
plot.show()
```

```python
#Exibindo informações estatisticas como: Total de componentes identificados e área dos objetos
(numCCs, imgLabels, stats, centroids) = cv2.connectedComponentsWithStats(img_bin, connectivity=4)
print("Total de Componentes Conexos encontrados: " + str(numCCs-1))
```

```python
#Exibindo os valores estatísticos do primeiro componente conexo detectado:
#Coordenadas X e Y
print("Primeiro píxel à esquerda: " + str(stats[1, [cv2.CC_STAT_LEFT]]))
print("Primeiro píxel no topo: " + str(stats[1, [cv2.CC_STAT_TOP]]))

#Informações sobre a Área do componente.
print("Valor de Largura: " + str(stats[1, [cv2.CC_STAT_WIDTH]]))
print("Valor de Alltura: " + str(stats[1, [cv2.CC_STAT_HEIGHT]]))
print("Valor de Área: " + str(stats[1, [cv2.CC_STAT_AREA]]))
```

```python
#Exibição da Imagem destacando os objetos conexos, cada objeto com uma cor diferente
plot.imshow(imgLabels, cmap='tab10');
```
