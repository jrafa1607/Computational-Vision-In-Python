```python
#Importando Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da imagem
img = cv2.imread(path + 'lenna.jpg')
```

```python
#Visualizao da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Converte-se a imagem colorida (BRG) para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Deteco de Bordas com Roberts
masc1 = np.array([[0, 1], [-1, 0]])
masc2 = np.array([[1, 0], [0, -1]])

img_roberts1 = cv2.filter2D(img_gray, -1, masc1)
img_roberts2 = cv2.filter2D(img_gray, -1, masc2)
img_roberts = cv2.add(img_roberts1, img_roberts2)

plot.imshow(img_roberts,cmap='gray')
plot.show()
```

```python
#Deteco de Bordas com Sobel
img_sobel1 = cv2.Sobel(img_gray,-1,1,0,ksize=3)
img_sobel2 = cv2.Sobel(img_gray,-1,0,1,ksize=3)
img_sobel = cv2.add(img_sobel1, img_sobel2)

plot.imshow(img_sobel,cmap='gray')
plot.show()
```

```python
#Deteco de Bordas com Prewitt
masc1 = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
masc2 = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])

img_prewitt1 = cv2.filter2D(img_gray, -1, masc1)
img_prewitt2 = cv2.filter2D(img_gray, -1, masc2)
img_prewitt = cv2.add(img_prewitt1, img_prewitt2)

plot.imshow(img_prewitt,cmap='gray')
plot.show()
```

```python
#Deteco de Bordas com Prewitt
img_canny = cv2.Canny(img_gray,100,200)

plot.imshow(img_canny,cmap='gray')
plot.show()
```
