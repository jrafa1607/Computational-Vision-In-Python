```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da Imagem
img = cv2.imread(path + 'lenna.jpg')
```

```python
#Visualizao da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Converte-se a imagem colorida (BRG) para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Para aplicar a transformao por potncia,  necessrio converter o intervalo de nveis de cinzas de imagem para [0,1]
#Descobre-se O Valor mximo de nveis de cinza
print("O Valor mximo de nveis de cinza : " + str(img_gray.max()))
```

```python
#Em seguida, dividi-se os valores da imagem em Grayscale pelo valor mximo de nvel de cinza
#Ser utilizado a letra E para a varivel Entrada, e a letra S para a varivel Sada

e = img_gray/img_gray.max()
```

```python
#Em seguida, defini-se um valor de gamma:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem  Clareada.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem  Escurecida.
    
#lembrando que quando uma imagem est no intervalo [0-255]:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem  Escurecida.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem  Clareada.
    
gamma = 10.5
```

```python
#Computando a imagem de sada, sendo que:
    #s = Imagem calculada
    #e = Escala da imagem grayscale dividida pelo valor mximo de nvel de cinza.
    #gamma = varivel de contraste
    
s = e ** gamma * 255
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
plot.show()
```

```python
#Transformao por Negativo
#Para aplicar a transformao por negativo, #basta descobrir qual o valor mximo da diviso entre:
    #A imagem em Grayscale / Valor mximo de nvel de cinza. Igual a transformao por potncia.

#Em ADIO,  necessrio subtrair o valor mximo desta diviso pelo seus valores de nvel de cinza.

e = img_gray/img_gray.max()
s = e.max()-e
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
```
