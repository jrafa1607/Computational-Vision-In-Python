```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da Imagem
img = cv2.imread(path + 'lenna.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Converte-se a imagem colorida (BRG) para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Para aplicar a transformação por potência, é necessário converter o intervalo de níveis de cinzas de imagem para [0,1]
#Descobre-se O Valor máximo de níveis de cinza
print("O Valor máximo de níveis de cinza é: " + str(img_gray.max()))
```

```python
#Em seguida, dividi-se os valores da imagem em Grayscale pelo valor máximo de nível de cinza
#Será utilizado a letra E para a variável Entrada, e a letra S para a variável Saída

e = img_gray/img_gray.max()
```

```python
#Em seguida, defini-se um valor de gamma:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem é Clareada.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem é Escurecida.
    
#lembrando que quando uma imagem está no intervalo [0-255]:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem é Escurecida.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem é Clareada.
    
gamma = 10.5
```

```python
#Computando a imagem de saída, sendo que:
    #s = Imagem calculada
    #e = Escala da imagem grayscale dividida pelo valor máximo de nível de cinza.
    #gamma = variável de contraste
    
s = e ** gamma * 255
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
plot.show()
```

```python
#Transformação por Negativo
#Para aplicar a transformação por negativo, #basta descobrir qual o valor máximo da divisão entre:
    #A imagem em Grayscale / Valor máximo de nível de cinza. Igual a transformação por potência.

#Em ADIÇÂO, é necessário subtrair o valor máximo desta divisão pelo seus valores de nível de cinza.

e = img_gray/img_gray.max()
s = e.max()-e
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
```
