```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se as imagens

img = cv2.imread('/content/A_mac_01.jpg')
template = cv2.imread('/content/T_mac.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Visualização do Template
plot.imshow(template,cmap='gray')
plot.show()
```

```python
#Coleta-se os valores de Altura e Largura do Template
h = template.shape[0]
w = template.shape[1]

print(h)
print(w)
```

```python
#Aplicação de Template Matching
res = cv2.matchTemplate(img,template,cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
```

<b> Outros Métodos de Similaridade que podem ser utilizados para Template Matching </b>

- cv.TM_CCOEFF
- cv.TM_CCOEFF_NORMED
- cv2.TM_CCORR
- cv2.TM_CCORR_NORMED
- cv2.TM_SQDIFF
- cv2.TM_SQDIFF_NORMED

```python
#Quando o método utilizado for SQDIFF Normalizado ou Não, não se utiliza o valor max_loc e sim o valor min_loc
#Para os demais, se utiliza o max_loc
top_left = max_loc
```

```python
#Em seguida, adiciona-se o objeto que irá destacar o template
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(img,top_left, bottom_right, (0,255,0), 2);
```

```python
#Exibição das Imagens: template
plot.imshow(template,cmap = 'gray')
plot.show()
```
