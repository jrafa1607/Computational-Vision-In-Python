```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
# Criando uma Função para aplicar Erosão e Dilatação
def erosao_dilatacao(img):

    # Defição de um kernel para as operações de erosão e dilatação
    kernel = np.ones((5, 5), np.uint8)  # Um kernel 5x5

    # Aplicando erosão
    imagem_erosionada = cv2.erode(img, kernel, iterations=1)

    # Aplicando dilatação
    imagem_dilatada = cv2.dilate(img, kernel, iterations=1)

    return imagem_erosionada, imagem_dilatada
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Importação + Conversão para Escala de Cinza

img = cv2.imread('letra_j_buraco.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
# Aplicação da Erosão e Dilatação
img_e, img_d = erosao_dilatacao(img)
```

```python
#Exibição dos Resultados
plot.figure(figsize=(10, 5))

plot.subplot(1, 3, 1)
plot.title("Imagem Original")
plot.imshow(img, cmap='gray')
plot.axis('off')

plot.subplot(1, 3, 2)
plot.title("Erosão")
plot.imshow(img_e, cmap='gray')
plot.axis('off')

plot.subplot(1, 3, 3)
plot.title("Dilatação")
plot.imshow(img_d, cmap='gray')
plot.axis('off')

plot.show()
```
