```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Importação + Conversão para Escala de Cinza
img = cv2.imread('Fingerprint.jpg', 0)
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Observação: Caso a imagem não esteja binarizada ainda, realize a conversão para sua Escala de Cinza, e posteriormente, sua Binarização
```

```python
#Em seguida, cria um EE de 3x3 com 1's
EE = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
```

```python
#Aplicação do Operador da Erosão
erosao = cv2.erode(img,EE,iterations = 1)
```

```python
#Aplicação do operador de Abertura na Imagem
abertura = cv2.morphologyEx(img, cv2.MORPH_OPEN, EE)
```

```python
#Aplicação do Operador de Dilatação na Abertura
dilatacao = cv2.dilate(abertura,EE,iterations = 1)
```

```python
#Aplicação do Operador de Fechamento na Abertura
fechamento = cv2.morphologyEx(abertura, cv2.MORPH_CLOSE, EE)
```

```python
#Exibição dos Resultados
fig = plot.figure(figsize=(30, 30))
plot.figsize=(50, 50)
ax1 = fig.add_subplot(1,5,1)
ax1.imshow(img,cmap = 'gray')
plot.title('Imagem com ruído')
ax2 = fig.add_subplot(1,5,2)
ax2.imshow(erosao,cmap = 'gray')
plot.title('Erosao da Imagem com Ruído')
ax3 = fig.add_subplot(1,5,3)
ax3.imshow(abertura,cmap = 'gray')
plot.title('Abertura da Erosão')
ax4 = fig.add_subplot(1,5,4)
ax4.imshow(dilatacao,cmap = 'gray')
plot.title('Dilatação pós Abertura')
ax4 = fig.add_subplot(1,5,5)
ax4.imshow(fechamento,cmap = 'gray')
plot.title('Fechamento pós Dilatação')
plot;
```

```python
#Resultados Imagem Original com Ruído x Imagem Após Pré-processamento com Erosão e Dilatação
fig = plot.figure(figsize=(15, 15))
plot.figsize=(15, 15)
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img,cmap = 'gray')
plot.title('Imagem Original com ruído')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(fechamento,cmap = 'gray')
plot.title('Imagem Pré-processada')
plot;
```
