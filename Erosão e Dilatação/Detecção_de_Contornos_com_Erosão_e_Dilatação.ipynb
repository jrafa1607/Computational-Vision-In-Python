```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Importação + Conversão para Escala de Cinza
img = cv2.imread('ShadowMan.png', 0)
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Em seguida, cria um EE de 5x5 com 1's
EE = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
```

```python
#Visualização do EE
print(EE)
```

```python
#Aplicação do Operador da Erosão
erosao = cv2.erode(img,EE,iterations = 1)
```

```python
#Computação do Contorno
contorno = img - erosao
```

```python
#Exibição dos Resultados
fig = plot.figure(figsize=(30, 30))
plot.figsize=(50, 50)
ax1 = fig.add_subplot(1,3,1)
ax1.imshow(img,cmap = 'gray')
plot.title('Imagem Original')
ax2 = fig.add_subplot(1,3,2)
ax2.imshow(erosao,cmap = 'gray')
plot.title('Erosão')
ax3 = fig.add_subplot(1,3,3)
ax3.imshow(contorno,cmap = 'gray')
plot.title('Contorno')
```

```python
#Visualização somente do Contorno
plot.imshow(contorno,cmap='gray')
plot.show()
```
