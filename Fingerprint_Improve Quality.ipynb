```python
#Import Libs
from matplotlib import pyplot as plot
import matplotlib as mpl
import urllib.request
import urllib.parse
import numpy as np
import random
import math
import cv2
```

```python
#Function to Load a Image from Github
def display_image_from_url(image_url):
    with urllib.request.urlopen(image_url) as url:
        s = url.read()
    img = cv2.imdecode(np.frombuffer(s, np.uint8), -1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
#Import Image
image_url = 'https://raw.githubusercontent.com/jrafa1607/Computational-Vision-In-Python/main/-%20Images/Fingerprint.jpg'
img = display_image_from_url(image_url)
```

```python
#Show the Image
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Applying Methods to Improve the Quality of the Fingerpint
EE = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
Erosion = cv2.erode(img,EE,iterations = 1)
Open = cv2.morphologyEx(img, cv2.MORPH_OPEN, EE)
Dilation = cv2.dilate(Open,EE,iterations = 1)
Close = cv2.morphologyEx(Open, cv2.MORPH_CLOSE, EE)
```

```python
# Set up the visualization
titles = ['Noise', 'Erosion', 'Open', 'Dilation', 'Close']
images = [img, Erosion, Open, Dilation, Close]
plot.figure(figsize=(13, 13))

for i in range(5):
    plot.subplot(1, 5, i+1)
    plot.imshow(images[i])
    plot.title(titles[i])
    plot.axis('off')

plot.tight_layout()
plot.show()
```

```python
#Final Result
fig = plot.figure(figsize=(15, 15))
plot.figsize=(15, 15)
ax1 = fig.add_subplot(1,2,1)
ax1.imshow(img,cmap = 'gray')
plot.title('Original Image with Noise')
ax2 = fig.add_subplot(1,2,2)
ax2.imshow(Close,cmap = 'gray')
plot.title('Final Image')
plot;
```
