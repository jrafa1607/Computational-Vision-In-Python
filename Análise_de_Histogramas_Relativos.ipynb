```python
#Importando Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/lenna.jpg')
```

```python
#Convers√£o da Imagem para escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Carrega-se o histograma da imagem bidimensional e converte-se essa entrada para unidimensional.
#A convers√£o dos dados de Matriz para vetor √© realizada por meio da fun√ß√£o ravel().

#Convers√£o dos dados Bidimensionais para Unidemensionais
histograma = plot.hist(img_gray.ravel(),256,[0,256])
np.set_printoptions(suppress=True)
```

```python
#Coleta-se as quantidades de Linhas e Colunas
num_lin = img_gray.shape[0]
num_col = img_gray.shape[1]

print(num_lin)
print(num_col)
```

```python
#Coleta-se o n√∫mero de pixels com o n√≠vel de cinza K [0 - 255]
H=histograma[0]
print(H)
```

```python
#Para calcular o Histograma Relativo, realiza-se a divis√£o do quantidade de pixels do n√≠vel k pelo total de pixels da imagem
#O total de pixels pode ser obtido multiplicando o total de linhas por colunas da imagem
total = num_lin * num_col
```

```python
#Exibi√ß√£o do Histograma Relativo
hr = H / total
print(hr)
```

```python
#Exibi√ß√£o do Histograma Relativo
plot.hist(hr.ravel())
plot.xlabel('N√≠veis de Cinza [√çndices]')
plot.ylabel('Quantidade de Pixels')
plot.title('Histograma Relativo')
plot.grid(True)
plot.show()
```
