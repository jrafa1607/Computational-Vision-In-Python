```python
#Import das Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/lenna.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Exemplos de Histogramas - Exibição das bandas R, G e B da Imagem
B, G, R = cv2.split(img)

#Banda Azul
hist_cv = cv2.calcHist([B],[0],None,[256],[0,256])
plot.plot(hist_cv, color='blue');

#Banda Verde
hist_cv = cv2.calcHist([G],[0],None,[256],[0,256])
plot.plot(hist_cv, color='green')

#Banda Vermelha
hist_cv = cv2.calcHist([R],[0],None,[256],[0,256])
plot.plot(hist_cv, color='red')
```

```python
#Converte-se a imagem colorida (BRG) para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Exibindo a o histograma com os níveis de cinza da Imagem
plot.title("Histograma Original")
plot.xlabel("Intensidade de Níveis de Cinza")
plot.ylabel("Quantidade de Pixels")
plot.hist(img_gray.ravel(), 256, [0,256], density=True)
plot.xlim([0, 256])
plot.grid(True)
plot.show()
```

```python
#Equaliza-se o histograma da imagem grayscale
img_histo_eq = cv2.equalizeHist(img_gray)
```

```python
#Exibição da Imagem Equalizada
plot.imshow(img_histo_eq,cmap='gray')
plot.show()
```

```python
#Exibição da Histograma da Imagem Equalizada
plot.title("Histograma Equalizado")
plot.xlabel("Intensidade de Níveis de Cinza")
plot.ylabel("Quantidade de Pixels")
plot.hist(img_histo_eq.ravel(), 256, [0,256], density=True)
plot.xlim([0, 256])
plot.grid(True)
plot.show()
```
