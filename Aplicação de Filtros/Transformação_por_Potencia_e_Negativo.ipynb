```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/lenna.jpg')
```

```python
#Conversão de BGR para RGB + Exibição da Imagem
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Converte-se a imagem colorida (BRG) para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**Aplicação de Transformação por Potência**

```python
#Para aplicar a transformação por potência, é necessário converter o intervalo de níveis de cinzas de imagem para [0,1]
#Inicialmente, descobre-se O Valor máximo de níveis de cinza
print(img_gray.max())
```

```python
#Em seguida, dividi-se os valores da imagem em Grayscale pelo valor máximo de nível de cinza
e = img_gray/img_gray.max()
```

```python
#Em seguida, defini-se um valor de gamma:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem é Clareada.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem é Escurecida.

#lembrando que quando uma imagem está no intervalo [0-255]:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem é Escurecida.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem é Clareada.

gamma = 6.5
```

```python
#Computando a imagem de saída
s = e ** gamma * 255
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
plot.show()
```

**Aplicação de Transformação por Negativo**

```python
#Inicialmente, descobre-se o valor da divisão entre a imagem em Grayscale / Valor máximo de nível de cinza
#Em seguida, é feito a subtração do valor desta divisão pelo seus valores de nível de cinza.

e = img_gray/img_gray.max()
s = e.max()-e
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
```
