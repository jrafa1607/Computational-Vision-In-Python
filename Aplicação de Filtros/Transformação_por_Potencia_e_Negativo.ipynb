```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/lenna.jpg')
```

```python
#Converso de BGR para RGB + Exibio da Imagem
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Converte-se a imagem colorida (BRG) para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

**Aplicao de Transformao por Potncia**

```python
#Para aplicar a transformao por potncia,  necessrio converter o intervalo de nveis de cinzas de imagem para [0,1]
#Inicialmente, descobre-se O Valor mximo de nveis de cinza
print(img_gray.max())
```

```python
#Em seguida, dividi-se os valores da imagem em Grayscale pelo valor mximo de nvel de cinza
e = img_gray/img_gray.max()
```

```python
#Em seguida, defini-se um valor de gamma:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem  Clareada.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem  Escurecida.

#lembrando que quando uma imagem est no intervalo [0-255]:
    #Se o valor de gamma for menor que 1 (< 1 ), a imagem  Escurecida.
    #Se o valor de gamma for maior que 1 (> 1 ), a imagem  Clareada.

gamma = 6.5
```

```python
#Computando a imagem de sada
s = e ** gamma * 255
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
plot.show()
```

**Aplicao de Transformao por Negativo**

```python
#Inicialmente, descobre-se o valor da diviso entre a imagem em Grayscale / Valor mximo de nvel de cinza
#Em seguida,  feito a subtrao do valor desta diviso pelo seus valores de nvel de cinza.

e = img_gray/img_gray.max()
s = e.max()-e
```

```python
#Exibindo os resultados
plot.imshow(e, cmap='gray')
plot.show()
plot.imshow(s, cmap='gray')
```
