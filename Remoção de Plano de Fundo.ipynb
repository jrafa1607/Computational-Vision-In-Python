```python
#Import das Libs
import cv2
import cvzone
import numpy as np
import matplotlib as mpl

from matplotlib import pyplot as plot
from cvzone.SelfiSegmentationModule import SelfiSegmentation;
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Background/'
```

```python
#Import da Imagem
#img = cv2.imread(path + 'flor.jpg')
img = cv2.imread(path + 'sunrise.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Converte-se a imagem colorida para Escalas de Cinza
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Criando o segmentador
segmentor = SelfiSegmentation()
```

```python
#Ajuste do tamanho da Imagem
img = cv2.resize(img, (640, 480))
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Em seguida, define-se um Limiar para separar Bits de cor preta (0) e Branca (255)
#Neste caso, definiu-se o valor de 125

T = 125
T ,img_bin = cv2.threshold(img, T, 255, cv2.THRESH_BINARY)
```

```python
#Visualização da Imagem Binarizada
plot.imshow(img_bin,cmap='gray')
plot.show()
```

```python
#Export da Imagem Binarizada
cv2.imwrite(path + 'sunrise_bin.jpg', img_bin)
```
