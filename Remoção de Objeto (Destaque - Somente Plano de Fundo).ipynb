```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Background/'
```

```python
#Import da Imagem
img = cv2.imread(path + 'flor.jpg')
```

```python
#Visualizao da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Coleta-se informaes de Tamanho da Imagem
hh, ww = img.shape[:2]
```

```python
#Definio do Limiar para Branco (Atribuio de novos Limites)
lower = np.array([185, 185, 185])
upper = np.array([255, 255, 255])
```

```python
#Criao da Mscara
thresh = cv2.inRange(img, lower, upper)
```

```python
#Visualizao da Mascara
plot.imshow(thresh,cmap='gray')
plot.show()
```

```python
#Aplicao da Morfologia
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (20,20))
morph = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel)
```

```python
#Inverso da Mascara
mask = 255 - morph
```

```python
#Resultado Final
result = cv2.bitwise_and(img, img, mask=mask)
plot.imshow(result,cmap='gray')
plot.show()
```

```python
#Export do Resultado Final
cv2.imwrite('D:/Google Drive/Github/CV/Background/flor_background.jpg', result)
```
