```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import imageio
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da Imagem
img = imageio.imread(path + 'lenna.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Leitura das colunas da imagem
fig, axs = plot.subplots(1, 4, sharey=True)
axs[0].imshow(img)
axs[1].imshow(img[:, ::-1, :])
axs[2].imshow(img[::-1, :, :])
axs[3].imshow(img[::-1, ::-1, :])
```

```python
#Atribuição das Imagens para Export
img2 = img[:, ::-1, :]
img3 = img[::-1, :, :]
img4 = img[::-1, ::-1, :]
```

```python
#Ordenando as Bandas RGB
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
cv2.imwrite(path + 'lenna-mirror.jpg', img2)
```

```python
#Ordenando as Bandas RGB
img3 = cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
cv2.imwrite(path + 'lenna-invert.jpg', img3)
```

```python
#Ordenando as Bandas RGB
img4 = cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
cv2.imwrite(path + 'lenna-mirror-invert.jpg', img4)
```
