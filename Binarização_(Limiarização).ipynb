```python
#Importa-se as Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem
img = cv2.imread('/content/lenna.jpg')
```

```python
#Inicialmente, converte-se a imagem colorida para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plot.imshow(img_gray,cmap='gray')
plot.show()
```

**Método 01 Binarização - Manualmente**

```python
#Cria-se uma matriz com as dimensões da imagem e define-se o limiar
img_bin=np.zeros((img.shape[0],img.shape[1]), dtype="uint8")
T = 85
```

```python
#Percorre-se os pixels da imagem, e se o pixel for maior que o Threshold, recebe 255, senão recebe 0
for l in range (img.shape[0]):
     for c in range (img.shape[1]):
         if(img_gray[l,c]>T):
             img_bin[l,c] = 255
         else: img_bin[l,c] = 0
```

```python
#Visualização das Imagens (Escala de Cinza e Binarizada)
plot.imshow(img_gray,cmap='gray')
plot.show()
plot.imshow(img_bin,cmap='gray')
plot.show()
```

**Método 02 Binarização - OpenCV + Função Thresh_Binary**

```python
#Define-se um valor de limiar, e em seguida, aplica-se a binarização
T = 105
T ,img_bin = cv2.threshold(img_gray, T, 255, cv2.THRESH_BINARY)
```

```python
#Visualização das Imagens (Escala de Cinza e Binarizada)
plot.imshow(img_gray,cmap='gray')
plot.show()
plot.imshow(img_bin,cmap='gray')
plot.show()
```

**Método 03 Binarização - OpenCV + Método OTSU**

```python
#Aplicação de OTSU com o OpenCV
otsu_threshold, img_bin = cv2.threshold(img_gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
#Visualização das Imagens (Escala de Cinza e Binarizada)
plot.imshow(img_gray,cmap='gray')
plot.show()
plot.imshow(img_bin,cmap='gray')
plot.show()
```
