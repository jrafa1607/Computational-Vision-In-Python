```python
#Import das Libs
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
#Import da Imagem do OneDrive
img = cv2.imread('D:/OneDrive/OneDrive - Etec Centro Paula Souza/Codes/Computational Vision/Imagens/lenna.jpg')
```

```python
#Separa-se as bandas em trÃªs imagens independentes
#Em segida, calcula-se e exibi-se os histogramas de cada banda.

B, G, R = cv2.split(img)
```

```python
#Banda Azul
hist_cv = cv2.calcHist([B],[0],None,[256],[0,256])
plt.plot(hist_cv, color='blue')
```

```python
#Banda Verde
hist_cv = cv2.calcHist([G],[0],None,[256],[0,256])
plt.plot(hist_cv, color='green')
```

```python
#Banda Vermelha
hist_cv = cv2.calcHist([R],[0],None,[256],[0,256])
plt.plot(hist_cv, color='red')
```

```python
#Converte-se a imagem colorida (BRG) para Escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Imagem Colorida', img)
cv2.imshow('Imagem Grayscale', img_gray);
cv2.waitKey(0);
```

```python
#Computar e Exibir o Histograma da Imagem Por meio da funÃ§Ã£o hist da biblioteca matplotlib
```

```python
#ExibiÃ§Ã£o do histograma usando Matplotlib
plt.hist(img_gray.ravel(),256,[0,256], density=True)
plt.xlabel('Quantidade de NÃ­veis de Cinza [Ãndices]')
plt.ylabel('Quantidade de Pixels')
plt.title('Histograma')
plt.grid(True)
plt.show()
```

```python
#cÃ¡lculo do histograma usando Numpy
hist_np,bins = np.histogram(img.ravel(),256,[0,256])
plt.plot(hist_np)
```

```python
#cÃ¡lculo do histograma usando OpenCV
hist_cv = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(hist_cv)
```

```python
#Calculando o Histograma Relativo
```

```python
#Inicialmente, recebe-se o histograma da imagem bidimensional e converte-se essa entrada em unidimensional.
#A conversÃ£o dos dados de entrada para vetor Ã© feito por meio da funÃ§Ã£o ravel().

histograma = plt.hist(img_gray.ravel(),256,[0,256])
np.set_printoptions(suppress=True)
print(histograma)
```

```python
#Coleta-se as quantidades de Linhas e Colunas
num_lin = img_gray.shape[0]
num_col = img_gray.shape[1]

print(num_lin)
print(num_col)
```

```python
#Coleta-se o nÃºmero de pixels com o nÃ­vel de cinza K [0 - 255, por exemplo]
#Quantos pixels com o nÃ­vel 0,1,2,3,4,5...255

H=histograma[0]
print(H)
```

```python
#Histograma Relativo
#Para calcular o Histograma Relativo, realiza-se a divisÃ£o do quantidade de pixels do nÃ­vel k pelo total de pixels da imagem
```

```python
#O total de pixels pode ser obtido multiplicando o total de linhas por colunas da imagem
total = num_lin * num_col
print(total)
```

```python
#No caso do histograma Relativo
hr = H / total
print(hr)
```

```python
#ExibiÃ§Ã£o do Histograma Relativo
plt.hist(hr.ravel())
plt.xlabel('NÃ­veis de Cinza [Ãndices]')
plt.ylabel('Quantidade de Pixels')
plt.title('Histograma Relativo')
plt.grid(True)
plt.show()
```
