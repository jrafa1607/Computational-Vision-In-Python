```python
#Importando Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da imagem
img = cv2.imread(path + 'lenna.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Exemplo de Histogramas com cada uma das bandas R, G e B da Imagem
#Inicialmente, separa-se a imagem em três bandas:
B, G, R = cv2.split(img)
```

```python
#Banda Azul
hist_cv = cv2.calcHist([B],[0],None,[256],[0,256])
plot.plot(hist_cv, color='blue');
```

```python
#Banda Verde
hist_cv = cv2.calcHist([G],[0],None,[256],[0,256])
plot.plot(hist_cv, color='green')
```

```python
#Banda Vermelha
hist_cv = cv2.calcHist([R],[0],None,[256],[0,256])
plot.plot(hist_cv, color='red')
```

```python
#Exemplos com a Imagem em Escala de Cinza
```

```python
#Conversão da Imagem para escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Visualização da Imagem
plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Calculando e Exibindo o Histograma da Imagem (Hist Function)
plot.hist(img_gray.ravel(),256,[0,256], density=True)
plot.xlabel('Quantidade de Níveis de Cinza')
plot.ylabel('Quantidade de Pixels')
plot.title('Histograma')
plot.grid(True)
plot.show()
```

```python
#Calculando e Exibindo o Histograma da Imagem (Numpy Function)
hist_np,bins = np.histogram(img.ravel(),256,[0,256])
plot.xlabel('Quantidade de Níveis de Cinza')
plot.ylabel('Quantidade de Pixels')
plot.title('Histograma')
plot.grid(True)
plot.plot(hist_np)
```

```python
#Calculando e Exibindo o Histograma da Imagem (OpenCV Function)
hist_cv = cv2.calcHist([img],[0],None,[256],[0,256])
plot.xlabel('Quantidade de Níveis de Cinza')
plot.ylabel('Quantidade de Pixels')
plot.title('Histograma')
plot.grid(True)
plot.plot(hist_cv)
```
