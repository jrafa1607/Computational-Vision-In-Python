```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import random
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da Imagem
img = cv2.imread(path + 'lenna.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Criando uma função para gerar ruídos aleatórios
def sp_noise(image,prob):

    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob 
    
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output
```

```python
#Atribuição do ruído na Imagem
noise_img = sp_noise(img,0.05)
```

```python
#Visualização da Imagem com Ruído
plot.imshow(noise_img,cmap='gray')
plot.show()
```

```python
#Export do Resultado Final
cv2.imwrite(path + 'lenna-salt-and-pepper.jpg', noise_img)
```
