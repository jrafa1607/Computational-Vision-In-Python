```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da Imagem
img = cv2.imread(path + 'lenna.jpg')
```

```python
#Visualizao da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Separao em trs bandas indenpendentes (RGB)
```

```python
#Mtodo 01 com Split
B, G, R = cv2.split(img)
```

```python
#Visualizao das Banbdas
plot.imshow(B,cmap='gray')
plot.show()
plot.imshow(G,cmap='gray')
plot.show()
plot.imshow(R,cmap='gray')
plot.show()
```

```python
#Mtodo 02 - Matrizes Indenpendentes
```

```python
#Extrai-se os valores de cada Matriz separadamente
B = img[:,:,0]
G = img[:,:,1]
R = img[:,:,2]
```

```python
#Visualizao das Banbdas
plot.imshow(B,cmap='gray')
plot.show()
plot.imshow(G,cmap='gray')
plot.show()
plot.imshow(R,cmap='gray')
plot.show()
```
