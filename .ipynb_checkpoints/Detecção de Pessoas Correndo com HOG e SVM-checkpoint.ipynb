```python
#Importando Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Run/'
```

```python
#Import da imagem
img = cv2.imread(path + 'run02.png')
```

```python
#Visualizao das Imagens
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Converso das Imagens para escalas de Cinza
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
#Visualizao das Imagens em Escala de Cinza
plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Carrega-se o Descritor de HOG 
hog = cv2.HOGDescriptor()
```

```python
#Para a Deteco de Pessoas com HOG, ser utilizado SVM
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
#Criando as caixas de Deteco para as Imagens
boxes, weights = hog.detectMultiScale(img_gray)
```

```python
#Novo Import da Imagem Colorida
imgColor = cv2.imread(path + "run02.png", cv2.IMREAD_COLOR)
plot.imshow(imgColor,cmap='gray')
plot.show()
```

```python
#Nova converso para escalas de Cinza
imgGray = cv2.cvtColor(imgColor, cv2.COLOR_BGR2GRAY)
plot.imshow(imgGray,cmap='gray')
plot.show()
```

```python
#Insero das Caixas na Imagem Colorida, aps a deteco do Descritor
for (x, y, w, h) in boxes:
    
    #Definio da Esquerda Superior, Direita Inferior, Cor e Largura da Caixa
    pontoEsqSuperior = (x, y)
    pontoDirInferior = (x + w, y + h)
    cor = (0, 255, 0)
    espessura = 2

    #Insero das Caixas
    cv2.rectangle(imgColor, pontoEsqSuperior, pontoDirInferior, cor, espessura)

plot.imshow(imgColor,cmap='gray')
plot.show()
```
