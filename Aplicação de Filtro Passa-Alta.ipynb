```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Path Definition:
path = 'D:/Google Drive/Github/CV/Lenna/'
```

```python
#Import da Imagem
img = cv2.imread(path + 'lenna.jpg')
```

```python
#Visualizao da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Aplicaao do filtro Passa-alta  utilizado para destacar bordas
#Para isso,  necessrio criar uma matriz 3x3, sendo que 8 unidades possuem o valor -1 e o centro possui o valor 8.

matriz = np.ones((3,3),np.float32)*-1
matriz[1][1] = 8
print(matriz)
```

```python
#Aplicao do Filtro
saida = cv2.filter2D(img,-1,matriz)
```

```python
#Visualizao da Imagem
plot.imshow(saida,cmap='gray')
plot.show()
```

```python
#Enfatizao das bordas
#Somatrio das imagens de entrada e sada.
image_final = saida + img
plot.imshow(image_final,cmap='gray')   
plot.show()
```

```python
#Export do Resultado Final
cv2.imwrite('D:/Google Drive/Github/CV/Lenna/bordas-enfatizadas.jpg', image_final)
```
