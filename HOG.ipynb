```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
PATH = "/content/drive/MyDrive/Colab Notebooks/Aulas/imagens"
```

```python
hog = cv.HOGDescriptor()
hog.setSVMDetector(cv.HOGDescriptor_getDefaultPeopleDetector())
```

```python
imgColor = cv.imread(PATH + "/pessoasCorrendo2.jpg", cv.IMREAD_COLOR)
imgGray = cv.cvtColor(imgColor, cv.COLOR_BGR2GRAY)

cv2_imshow(imgColor) 
```

```python
boxes, weights = hog.detectMultiScale(imgGray )

for (x, y, w, h) in boxes:
  pontoEsqSuperior = (x, y)
  pontoDirInferior = (x + w, y + h)
  cor = (0, 255, 0)
  espessura = 2

  cv.rectangle(imgColor, pontoEsqSuperior, pontoDirInferior, cor, espessura)

cv2_imshow(imgColor) 
```
