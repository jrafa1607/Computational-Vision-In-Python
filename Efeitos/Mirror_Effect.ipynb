```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/lenna.jpg')
```

```python
#Conversão de BGR para RGB + Exibição da Imagem
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Matriz 4x1
fig, axs = plot.subplots(1, 4, sharey=True)
axs[0].imshow(img)
axs[1].imshow(img[:, ::-1, :])
axs[2].imshow(img[::-1, :, :])
axs[3].imshow(img[::-1, ::-1, :])
```
