```python
#Importando Libs
import numpy as np
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.\n"
#Em seguida, importa-se a imagem\n"
img = cv2.imread('/content/lenna.jpg')
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Conversão de BGR para RGB
img_rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img_rgb,cmap='gray')
plot.show()
```

```python
#Conversão de BGR para Gray Scale
img_gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plot.imshow(img_gray,cmap='gray')
plot.show()
```

```python
#Conversão de BGR para Gray Scale Manualmente
img_gray=np.zeros((img.shape[0],img.shape[1]), dtype="uint8")
```

```python
#Cria-se laços para percorrer os pixels de linha e coluna;
#Em seguida, para cada pixel [l,c], soma-se as 3 camadas (RGB) e dividi-se 3
#Com isso, obtêm-se uma única camada em escala de cinza

for l in range (img.shape[0]):
     for c in range (img.shape[1]):
         (b, g, r) = img[l,c]
         img_gray[l,c] = (int(b)+int(g)+int(r))//3
```

```python
#Visualização da Imagem
plot.imshow(img,cmap='gray')
plot.show()
plot.imshow(img_gray,cmap='gray')
plot.show()
```
