```python
#Import das Libs
import matplotlib as mpl
import numpy as np
import math
import cv2
from matplotlib import pyplot as plot
```

```python
#Realiza-se o Upload da Imagem para o Google Colab.
#Em seguida, importa-se a imagem

img = cv2.imread('/content/lenna.jpg')
```

```python
#Conversão de BGR para RGB + Exibição da Imagem
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plot.imshow(img,cmap='gray')
plot.show()
```

**Separação de uma Imagem em três bandas indenpendentes (R, G e B)**

* Método 01 - Utilizando Split

```python
#Aplicação do Split
B, G, R = cv2.split(img)

#Visualização das Banbdas
plot.imshow(B,cmap='gray')
plot.show()
plot.imshow(G,cmap='gray')
plot.show()
plot.imshow(R,cmap='gray')
plot.show()
```

* Método 02 - Utilizando Matrizes Indenpendentes

```python
#Extrai-se os valores de cada Matriz separadamente
B = img[:,:,0]
G = img[:,:,1]
R = img[:,:,2]

#Visualização das Banbdas
plot.imshow(B,cmap='gray')
plot.show()
plot.imshow(G,cmap='gray')
plot.show()
plot.imshow(R,cmap='gray')
plot.show()
```
