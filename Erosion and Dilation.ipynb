```python
#Import Libs
from matplotlib import pyplot as plot
import matplotlib as mpl
import urllib.request
import urllib.parse
import numpy as np
import random
import math
import cv2
```

```python
#Function to Load a Image from Github
def display_image_from_url(image_url):
    with urllib.request.urlopen(image_url) as url:
        s = url.read()
    img = cv2.imdecode(np.frombuffer(s, np.uint8), -1)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img
```

```python
#Import Image
image_url = 'https://raw.githubusercontent.com/jrafa1607/Computational-Vision-In-Python/main/-%20Images/ShadowMan.png'
img = display_image_from_url(image_url)
```

```python
#Show the Image
plot.imshow(img,cmap='gray')
plot.show()
```

### **Erosion and Dilation**

**Erosion** and **Dilation** are fundamental morphological operations in image processing, often used for noise reduction, feature extraction, and image segmentation. These operations work by applying a structuring element (kernel) to an input image.


*   **Erosion:** Erosion shrinks or thins the foreground object (usually white pixels) in a binary image. Conceptually, it's like 'eroding away' the boundaries of foreground objects. For each pixel in the image, the structuring element is centered on that pixel. If and only if **all** pixels under the structuring element are foreground (white), then the central pixel of the output image remains foreground. Otherwise, it becomes background (black). The applications of Erosion Technique include: **Noise removal**, **Disconnecting objects:** and **Thinning**.

*   **Dilation:** Dilation expands or thickens the foreground object in a binary image. It's the opposite of erosion, 'growing' the boundaries of foreground objects. For each pixel in the image, the structuring element is centered on that pixel. If *at least one* pixel under the structuring element is foreground (white), then the central pixel of the output image becomes foreground. Otherwise, it remains background. The applications of Dilation include: **Filling holes**, **Connecting objects** and **Thickening**.

* **Important:** Both erosion and dilation use a **structuring element** (also known as a kernel). This is a small matrix (e.g., 3x3 or 5x5) that defines the neighborhood of pixels to be considered for the operation. The shape and size of the structuring element determine the nature and extent of the morphological operation.

```python
#Create a EE of 5x5
EE = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
```

```python
#Create a Erosion Opeartor
Erosion = cv2.erode(img,EE,iterations = 1)
```

```python
#Show the Shape
Shape = img - Erosion
```

```python
# Set up the visualization
titles = ['Original', 'Erosion', 'Shape']
images = [img, Erosion, Shape]
plot.figure(figsize=(7, 7))

for i in range(3):
    plot.subplot(1, 3, i+1)
    plot.imshow(images[i])
    plot.title(titles[i])
    plot.axis('off')

plot.tight_layout()
plot.show()
```

```python
#Example - Filling Holes
#Import Image
image_url = 'https://raw.githubusercontent.com/jrafa1607/Computational-Vision-In-Python/main/-%20Images/letter_j.png'
img = display_image_from_url(image_url)
```

```python
#Show the Image
plot.imshow(img,cmap='gray')
plot.show()
```

```python
#Create a Function to Apply: Erosion and Dilation
def erosion_dilation(img):

    kernel = np.ones((5, 5), np.uint8)  # kernel 5x5
    img_erosion = cv2.erode(img, kernel, iterations=1)
    img_dilation = cv2.dilate(img, kernel, iterations=1)

    return img_erosion, img_dilation
```

```python
# Apply the Erosion and Dilation
img_e, img_d = erosion_dilation(img)
```

```python
# Set up the visualization
titles = ['Original', 'Erosion', 'Dilation']
images = [img, img_e, img_d]
plot.figure(figsize=(7, 7))

for i in range(3):
    plot.subplot(1, 3, i+1)
    plot.imshow(images[i])
    plot.title(titles[i])
    plot.axis('off')

plot.tight_layout()
plot.show()
```
